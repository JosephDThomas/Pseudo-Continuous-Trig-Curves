```python
import math
import random as r
import numpy as np
import sys
import time

np.set_printoptions(threshold=sys.maxsize)

print("Please input what type of curve you want to make psuedo-contiuous")
print("input 'sin' for a sine curve and 'cos' for a cosine curve")

curve = input()

if curve == "sin":
    print("a sine curve it is then!")
elif curve == "cos":
    print("a cosine curve it is then!")
else:
    print("this is not a correct input, please try again")
    sys.exit

print("please input the amplitude of the required wave")

amplitude = input()

print("Okay an amplitude of " + amplitude)

#parameters
dr = amplitude
points_x= 1000
points_y = 1000

#Starting empty matrix
A = np.zeros((points_x, points_y))

#Setting start postions
A[points_y-1,0] = 1
A[points_y-1, points_x-1] = 1
A[points_y-2,1] = 1
A[points_y-2, points_x-2] = 1

n = 1 #Iteration count
m = points_y-2 #Have to work backwards in y given that an element in A is set as Ayx in this code to get the pseudo-sine curve drawn in the matrix 

#Iterating across the grid 
if curve == "cos":
    for i in range(0,len(A)-1):
    #the three possible movements errors
        a = (np.cos(i*dr)-(m+1)*dr)**2

        b = (np.cos(i*dr)-(m)*dr)**2

        c = (np.cos(i*dr)-(m-1)*dr)**2

        choice = min(a,b,c)
        if choice == a:
            A[i+1,m+1] = 1
        elif choice == b:
            A[i+1,m] = 1
        else:
            A[i+1,m-1] = 1
    
    m = m - 1
elif curve == "sin":
    for i in range(0,len(A)-1):
    #the three possible movements errors
        a = (np.sin(i*dr)-(m+1)*dr)**2

        b = (np.sin(i*dr)-(m)*dr)**2

        c = (np.sin(i*dr)-(m-1)*dr)**2

        choice = min(a,b,c)
        if choice == a:
            A[i+1,m+1] = 1
        elif choice == b:
            A[i+1,m] = 1
        else:
            A[i+1,m-1] = 1
    
    m = m - 1

    #iteration count
    n = n+1
    if n%100 == 0:
        print(n)

print("it's done!")

```
